```python
#https://python.tutorialink.com/how-can-i-split-an-audio-file-into-multiple-audio-wav-files-from-folder/
from pydub import AudioSegment 
from pydub.utils import make_chunks
import os
```

```python
def process_sudio(file_name, directory):
    myaudio = AudioSegment.from_file(file_name, "wav") 
    chunk_length_ms = 10000 # pydub calculates in millisec 
    chunks = make_chunks(myaudio, chunk_length_ms) #Make chunks of ten sec 
    for i, chunk in enumerate(chunks): 
        chunk_name =  r'C:\Users\jemilad\OneDrive - SINTEF\Desktop\audio\\' + directory + '\\' + each_file + "_{0}.wav".format(i)
        print ("exporting", chunk_name) 
        chunk.export(chunk_name, format="wav") 
```

```python
path_merd3 = 'D:\Master\Rataren2021\Merd3-etteravlusing'
directory_merd3 = 'merd3_chunked'

all_file_names = os.listdir(path_merd3)
try:
    os.makedirs(directory_merd3) # creating a folder named merd3_chunked
except:
    pass
for each_file in all_file_names:
    if ('.wav' in each_file):
        process_sudio(path_merd3 + '\\' + each_file, directory_merd3)
```
